```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
# **تحميل وتحضير بيانات MNIST**
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# تطبيع البيانات إلى نطاق [0, 1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# إعادة تشكيل البيانات لتناسب شبكة CNN
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# تحويل الصور إلى 3 قنوات لتناسب MobileNetV2
x_train = np.repeat(x_train, 3, axis=-1)
x_test = np.repeat(x_test, 3, axis=-1)

# **تغيير حجم الصور إلى 32x32 وضمان أن البيانات صحيحة**
x_train = tf.image.resize(x_train, (32, 32))
x_test = tf.image.resize(x_test, (32, 32))

# **تحويل البيانات إلى NumPy arrays والتأكد من نوعها**
x_train = np.array(x_train, dtype=np.float32)
x_test = np.array(x_test, dtype=np.float32)

# **التأكد من أن الأشكال متوافقة مع النموذج**
print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)

# تحويل التصنيفات إلى one-hot encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# **تحميل نموذج MobileNetV2 المسبق التدريب**
base_model = keras.applications.MobileNetV2(input_shape=(32, 32, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # تجميد الطبقات الأساسية

# **بناء نموذج Fine-Tuning**
model = keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(10, activation="softmax")
])

# **تجميع النموذج**
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# **تدريب النموذج الأساسي**
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

# **إلغاء تجميد بعض الطبقات لعمل Fine-Tuning**
base_model.trainable = True
for layer in base_model.layers[:100]:
    layer.trainable = False

# **إعادة تجميع النموذج بمعدل تعلم أقل**
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-5),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# **تدريب النموذج بعد Fine-Tuning**
history_finetune = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

# **تقييم النموذج**
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# **رسم منحنى الأداء**
plt.plot(history.history['accuracy'], label='Initial Training Accuracy')
plt.plot(history_finetune.history['accuracy'], label='Fine-Tuning Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
